adding columns for Total score and average

```python
df['Total_score']=df['math score'] +df['reading score']+df['writing score']
df['Total_score']
df['Average_score']=df['Total_score']/3
```

```python
df.head(2)
```

```python
reading_full=df[df['reading score']==100]['Average_score'].count()
writing_full=df[df['writing score']==100]['Average_score'].count()
math_full=df[df['math score']==100]['Average_score'].count()

print(f" number of students with full marks in maths are {math_full}")
print(f"number pf students with full marks in writing are {writing_full}")
print(f"number of students with full reading marks are {reading_full}")
```

```python
reading_full=df[df['reading score']<=20]['Average_score'].count()
writing_full=df[df['writing score']<=20]['Average_score'].count()
math_full=df[df['math score']<=20]['Average_score'].count()

print(f"number of students with less than 20 marks in maths are {math_full}")
print(f"number of students with mark less than 20 in reading are {reading_full}")
print(f"number of students with less than 20 marks in writing are {writing_full}")
```

INSIGHTS
* From above we can say that students have performed well in Writing  and then  Reading
* Best performance is in Writing section 
* They need to improve in Maths subject
